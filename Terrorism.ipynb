### **Printing first 5 rows**

```sql
SELECT TOP 5 *
FROM master.dbo.combined_terrorist
```

### **Rolling sum of deaths in an attack**

```sql

SELECT row_num, iyear, nkill
, SUM(nkill) OVER (ORDER BY row_num) AS total_nkill
FROM master.dbo.combined_terrorist
```

### **Deaths by Decade**

```sql
DROP TABLE IF EXISTS #decades
SELECT *,
CASE
WHEN iyear LIKE '197%' THEN 0
WHEN iyear LIKE '198%' THEN 1
WHEN iyear LIKE '199%' THEN 2
WHEN iyear LIKE '200%' THEN 3
WHEN iyear LIKE '201%' THEN 4
 
END AS decades
 
INTO #decades
FROM master.dbo.combined_terrorist
```

```sql
SELECT TOP 5 iyear, decades FROM #decades
```
